```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load dataset
df = pd.read_csv("../dataset/paygap.csv")

# show first rows
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["diff_mean_hourly_percent"], bins=30, kde=True)

plt.title("Distribution of Gender Pay Gap")
plt.xlabel("Mean Hourly Pay Gap (%)")
plt.ylabel("Number of Companies")

plt.show()
```

```python
top_companies = df.sort_values("diff_mean_hourly_percent", ascending=False).head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    x="diff_mean_hourly_percent",
    y="employer_name",
    data=top_companies
)

plt.title("Top 10 Companies with Highest Gender Pay Gap")
plt.xlabel("Pay Gap (%)")
plt.ylabel("Company")

plt.show()

```

```python
bonus_avg = df[["male_bonus_percent","female_bonus_percent"]].mean()

plt.figure(figsize=(6,4))

bonus_avg.plot(kind="bar")

plt.title("Average Bonus Percentage by Gender")
plt.ylabel("Bonus Percentage")
plt.xticks(rotation=0)

plt.show()
```

```python
quartile_data = df[
[
"male_lower_quartile",
"female_lower_quartile",
"male_upper_middle_quartile",
"female_upper_middle_quartile",
"male_top_quartile",
"female_top_quartile"
]
].mean()

plt.figure(figsize=(10,5))

quartile_data.plot(kind="bar")

plt.title("Gender Distribution Across Pay Levels")
plt.ylabel("Percentage")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
df[
[
"diff_mean_hourly_percent",
"diff_median_hourly_percent",
"male_bonus_percent",
"female_bonus_percent"
]
].corr(),
annot=True,
cmap="coolwarm"
)

plt.title("Correlation Between Pay Gap Variables")

plt.show()
```
